```python
import numpy as np
import matplotlib.pyplot as plt
import os
#with OS package, use /Users/username instead of ~ for the home directory
#pip install opencv-python
import cv2

import random
import collections
import pickle
```

```python
PATH = '/Users/edwinsalguero/Desktop/Edwin/Academic/Summer2022/MSDS434/Project/archive/images'

TRAIN_PATH = PATH + '/images'
#TEST_PATH = PATH + '/test'

CATEGORIES = ['AIRPLANE']

#print(os.getcwd())
```

```python
def getImages( WHERE, IMG_SIZE ) :
    imgList = []
    for category in CATEGORIES :
        target_num = CATEGORIES.index( category )
        path = os.path.join( WHERE, category )
        print( "path=", path)
        for img in os.listdir( path ) :
            try:
                filename = os.path.join( path, img )
                img_array = cv2.imread( filename, cv2.IMREAD_GRAYSCALE )
                print( "BEFORE .... ", img_array.shape )
                new_array = cv2.resize( img_array, (IMG_SIZE, IMG_SIZE) )
                print( "AFTER .... ", new_array.shape )
                imgList.append( [new_array, target_num])
            except Exception as e:
                pass
    return imgList
```

```python
# SET THIS TO WHATEVER SIZE YOU THINK IS BEST ....
IMG_SIZE = 10

# trainData = getImages( TEST_PATH, IMG_SIZE  )
# testData = getImages( TEST_PATH, IMG_SIZE  )

#totalData = getImages( TRAIN_PATH, IMG_SIZE  ) + getImages( TEST_PATH, IMG_SIZE  )
totalData = getImages( PATH, IMG_SIZE  ) 
```

```python
who = 0

print( "Total Images",len( totalData ))
countList = [i[1] for i in totalData ]
ctr = collections.Counter( countList )
for i in CATEGORIES :
    print("Category",i,"=",ctr[ CATEGORIES.index(i)] )

random.shuffle( totalData )
theRecord = totalData[who]
theValue = theRecord[1]
theImg = theRecord[0]
plt.imshow( theImg, cmap="gray" )
plt.show()

print("This is a",CATEGORIES[ theValue ] )
```

```python
# 10 x 10 grid, each pixel has a single value associated with it
X = []
Y = []

for [features,label] in totalData :
    X.append( features )
    Y.append( label )
#for a color image you would use (-1, IMG_SIZE, IMG_SIZE, 3) the last position is the number of channels
# the negative is useful to interpret the correct number of images and input it in where the negative one is
X = np.array( X ).reshape( -1, IMG_SIZE, IMG_SIZE, 1)
print( X.shape )

Y = np.array( Y ).reshape( -1, )
print( Y.shape )
```

```python
pickle_out = open("X.pickle","wb")
pickle.dump( X, pickle_out )
pickle_out.close()

pickle_out = open("Y.pickle","wb")
pickle.dump( Y, pickle_out )
pickle_out.close()
```
